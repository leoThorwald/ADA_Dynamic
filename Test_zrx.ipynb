```python
import pandas as pd
import re
import json
```

```python
movie = pd.read_csv("MovieSummaries/movie.metadata.tsv", delimiter='\t',header=None)
character = pd.read_csv("MovieSummaries/character.metadata.tsv", delimiter='\t',header=None)
```

# Add column names

```python
character.columns = ['Wikipedia Movie ID','Freebase Movie ID','Actor DOD','Character Name','Actor DOB','Actor gender','Actor height','Actor ethnicity','Actor Name','Actor age at movie release','Freebase character map1','Freebase character map2','Freebase character map3']
character
```

```python
movie.columns = ['Wikipedia movie ID','Freebase movie ID','Movie name','Movie release date','Movie box office revenue','Movie runtime','Movie languages','Movie countries','Movie genres']
movie
```

# Text regularization

```python
movie.dtypes
```

```python
character.dtypes
```

```python
def extract_after(text,keyword):
    match = re.compile(rf"{keyword}(.*)", re.IGNORECASE).search(text)
    return match.group(1).strip() if match else ""
def extract_before(text,keyword):
    match = re.search(rf'^(.*?)(?={re.escape(keyword)})', text)
    return match.group(1).strip() if match else text
def extract_between(text,start,end):
    match = re.compile(rf"{re.escape(start)}(.*?){re.escape(end)}", re.IGNORECASE).search(text)
    return match.group(1).strip() if match else ""
```

```python
movie['Movie release date'] = movie['Movie release date'].apply(lambda x: extract_before(str(x), '-'))
movie['Movie release date'] = pd.to_datetime(movie['Movie release date'], errors='coerce').dt.year
#movie['Freebase movie ID'] = movie['Freebase movie ID'].apply(lambda x: extract_after(x, '/m/'))
movie['Movie languages'] = movie['Movie languages'].apply(lambda x: extract_between(x, '": "', ' Language'))
movie['Movie countries'] = movie['Movie countries'].apply(lambda x: extract_between(x, '": "', '"}'))
#movie['Movie languages'] = movie['Movie languages'].apply(lambda x: list(json.loads(x).values()))
#movie['Movie countries'] = movie['Movie countries'].apply(lambda x: list(json.loads(x).values()))
movie['Movie genres'] = movie['Movie genres'].apply(lambda x: list(json.loads(x).values()))
movie
```
